### Task: Prediction Using Decision Tree Algorithm

#### 1. Importing libararies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import f1_score, hinge_loss, confusion_matrix
from sklearn.datasets import load_iris

%matplotlib inline
```

#### 2.  Loading & Converting Dataset

```python
iris=load_iris()
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df.describe()
```

```python
df['target']=iris.target
df.sample(5)
```

#### 3. Train-Test Split

```python
x_train, x_test, y_train, y_test= train_test_split(
                                                   df.drop('target',axis=1), df['target'] ,test_size=.2 ) 
```

#### 4. Building Classifier

```python
ds_tree = DecisionTreeClassifier()
ds_tree.fit(x_train, y_train)
```

#### 5. Plotting Decision Tree by plot_tree function

```python
plt.figure(figsize=(20,20))
plot_tree(ds_tree, filled=True) #For Plotting Decision 
plt.show()
```

#### 6. Checking Accuracy & Plotting Confusion Matrix for MissClassified Point

```python
print( f"Accuracy: {ds_tree.score(x_train, y_train) * 100} %" )
prd_y = ds_tree.predict(x_test)
sns.heatmap(confusion_matrix(y_test,prd_y) ,annot=True)
plt.show()
```

### Conclusion: Only 3 Flowers are Miss-Classified
